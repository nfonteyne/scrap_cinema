```python
def extract_movie_data(div):
    movie = {}
    
    try:
        movie['title'] = div.find('a', class_='meta-title-link').text.strip()
    except AttributeError:
        movie['title'] = 'N/A'
    
    try:
        movie['release_date'] = div.find('span', class_='date').text.strip()
    except AttributeError:
        movie['release_date'] = 'N/A'
    
    try:
        movie['genre'] = div.find('span', class_='dark-grey-link').text.strip()
    except AttributeError:
        movie['genre'] = 'N/A'
    
    movie['nationalities'] = [span.text.strip() for span in div.find_all('span', class_='nationality')]
    
    try:
        movie['director'] = div.find('div', class_='meta-body-direction').find('span', class_='dark-grey-link').text.strip()
    except AttributeError:
        movie['director'] = 'N/A'
    
    try:
        actors_div = div.find('div', class_='meta-body-item meta-body-actor')
        if actors_div:
            actors_spans = actors_div.find_all('span', class_='dark-grey-link')
            movie['actors'] = [span.text.strip() for span in actors_spans]
        else:
            movie['actors'] = []
    except AttributeError:
        movie['actors'] = []
 
    try:
        press_rating_elem = div.find('span', text=' Presse ')
        movie['press_rating'] = press_rating_elem.find_next('span', class_='stareval-note').text.strip() if press_rating_elem else 'N/A'
    except AttributeError:
        movie['press_rating'] = 'N/A'
    
    try:
        spectator_rating_elem = div.find('span', text=' Spectateurs ')
        movie['spectator_rating'] = spectator_rating_elem.find_next('span', class_='stareval-note').text.strip() if spectator_rating_elem else 'N/A'
    except AttributeError:
        movie['spectator_rating'] = 'N/A'
    
    try:
        movie['synopsis'] = div.find('div', class_='content-txt').text.strip()
    except AttributeError:
        movie['synopsis'] = 'N/A'
    
    movie['showtimes'] = [span.find('span', class_='showtimes-hour-item-value').text.strip() 
                          for span in div.find_all('span', class_='showtimes-hour-item') if span.find('span', class_='showtimes-hour-item-value')]
    
    return movie
```

```python
url = "https://www.allocine.fr/seance/salle_gen_csalle=C0071.html#shwt_date=2024-10-24"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
# Use the function
movie_divs = soup.find_all('div', class_='card entity-card entity-card-list movie-card-theater cf hred')
movies = [extract_movie_data(div) for div in movie_divs]
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# Fonction pour extraire les données du film
def extract_movie_data(div):
    movie = {}
    
    try:
        movie['title'] = div.find('a', class_='meta-title-link').text.strip()
    except AttributeError:
        movie['title'] = 'N/A'
    
    try:
        movie['release_date'] = div.find('span', class_='date').text.strip()
    except AttributeError:
        movie['release_date'] = 'N/A'
    
    try:
        # Le genre peut être contenu dans le premier span sans classe spécifique
        genre_info = div.find('div', class_='meta-body-info')
        genres = genre_info.find_all('span', class_=False)
        movie['genre'] = ', '.join([g.text.strip() for g in genres])
    except AttributeError:
        movie['genre'] = 'N/A'
    
    movie['nationalities'] = [span.text.strip() for span in div.find_all('span', class_='nationality')]
    
    try:
        directors = div.find('div', class_='meta-body-direction').find_all('a', class_='blue-link')
        movie['director'] = ', '.join([d.text.strip() for d in directors])
    except AttributeError:
        movie['director'] = 'N/A'
    
    try:
        actors_div = div.find('div', class_='meta-body-actor')
        if actors_div:
            actors_spans = actors_div.find_all('a', class_='blue-link')
            movie['actors'] = [span.text.strip() for span in actors_spans]
        else:
            movie['actors'] = []
    except AttributeError:
        movie['actors'] = []
 
    try:
        # Extraire la note de presse
        press_rating_elem = div.find('span', class_='stareval-note', text=True)
        movie['press_rating'] = press_rating_elem.text.strip() if press_rating_elem else 'N/A'
    except AttributeError:
        movie['press_rating'] = 'N/A'
    
    try:
        # Extraire la note des spectateurs
        spectator_rating_elem = div.find('div', class_='rating-item')
        if spectator_rating_elem:
            spectator_note = spectator_rating_elem.find('span', class_='stareval-note')
            movie['spectator_rating'] = spectator_note.text.strip() if spectator_note else 'N/A'
        else:
            movie['spectator_rating'] = 'N/A'
    except AttributeError:
        movie['spectator_rating'] = 'N/A'
    
    try:
        movie['synopsis'] = div.find('div', class_='content-txt').text.strip()
    except AttributeError:
        movie['synopsis'] = 'N/A'
    
    try:
        # Extraire les horaires des séances
        showtimes_section = div.find('div', class_='showtimes-list')
        if showtimes_section:
            times = showtimes_section.find_all('a', class_='showtimes-hour-item')
            movie['showtimes'] = [time.text.strip() for time in times]
        else:
            movie['showtimes'] = []
    except AttributeError:
        movie['showtimes'] = []
    
    return movie

# Configurer Selenium pour utiliser Chrome en mode sans tête
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--no-sandbox')

# Initialiser le service Chromedriver
service = Service(ChromeDriverManager().install())

# Initialiser le WebDriver avec le service et les options
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL de la page à scraper
url = "https://www.allocine.fr/seance/salle_gen_csalle=C0071.html#shwt_date=2024-10-24"

# Naviguer vers l'URL
driver.get(url)

# Attendre que le contenu soit chargé
time.sleep(5)  # Vous pouvez ajuster la durée selon votre connexion

# Récupérer le HTML rendu
html = driver.page_source

# Fermer le navigateur
driver.quit()

# Utiliser BeautifulSoup pour analyser le HTML
soup = BeautifulSoup(html, 'html.parser')

# Trouver tous les divs contenant les films
movie_divs = soup.find_all('div', class_='card movie-card cf')

# Extraire les données des films
movies = [extract_movie_data(div) for div in movie_divs]

# Afficher les résultats
for movie in movies:
    print(movie)
```
